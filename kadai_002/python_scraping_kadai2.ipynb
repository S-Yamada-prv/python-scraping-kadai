```python
import requests
import pprint
from getpass import getpass

# APIキーを指定
api_key = getpass('APIキーを入力してください: ')
```

```python
# 駅名を指定
search_station = input('探索する駅名を入力してください: ')
```

```python
# 指定された駅の緯度経度を取得する
map_url = 'https://maps.googleapis.com/maps/api/geocode/json'

# データ取得リクエストの実行
response = requests.get(
    map_url,
    params={
        'key': api_key,  # APIキーをパラメータに追加
        'address': search_station  # 駅名
    }
)

# レスポンスを取得
json_data = response.json()
# 取得内容を確認する
pprint.pprint(json_data)
```

```python
import time

res_url = 'https://maps.googleapis.com/maps/api/place/nearbysearch/json'
for res in json_data['results']:
  # レストラン情報を取得して出力する
  rest_response = requests.get(
      res_url,
      params={
        'key': api_key,       # APIキー
        'location': str(res['geometry']['location']['lat']) + ',' + str(res['geometry']['location']['lng']),      # 緯度経度
        'language': 'ja',     # 日本語
        'radius': 500,        # 検索半径
        'type': 'restaurant'  # レストラン情報
      }
  )
  # レスポンスを取得
  rest_json_data = rest_response.json()
  # 複数ページの取得が必要な場合があるので無限ループにしておく
  while 1:
    # 1ページ分のレストラン情報を取得して処理する
    for rest_res in rest_json_data['results']:
      # 店名・評価・住所を出力する
      print(f"「店名：{rest_res['name']}」「評価：{rest_res['rating']}」「住所：{rest_res['vicinity']}」")
    # 次ページがある場合は取得する
    if ('next_page_token' in rest_json_data) and (rest_json_data['next_page_token'] != ''):
      nextpage_token = rest_json_data['next_page_token']
      time.sleep(2)
      rest_response = requests.get(
          res_url,
          params={
              'key': api_key,
              'pagetoken': nextpage_token
          }
      )
      rest_json_data = rest_response.json()
    # 最終ページを終了したらループを抜ける
    else:
      break
```
