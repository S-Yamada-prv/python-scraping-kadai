```python
import requests
from bs4 import BeautifulSoup

# タイトルを取得する対象のURL
url = 'https://news.yahoo.co.jp/articles/9a218a3dd49d789b50a43c9550e44fbffec3e85e'

# GETリクエストを送信し、HTMLを取得する
response = requests.get(url)

# HTMLを解析する
soup = BeautifulSoup(response.text, 'html.parser')

# CSSセレクタを使用して、ニュース本文を取得する
article_element = soup.select_one('#uamods > .article_body > .sc-zniwbk-0 > .sc-54nboa-0')
article_text = article_element.text
print(article_text)
```
